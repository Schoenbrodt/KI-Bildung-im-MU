# Lineare Regression


Zunächst werden relevante Pakete geladen.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

Im Folgenden Code werden zunächst zweidimensionale Datenpunkte erzeugt. Anschließend wird die optimale Regressionagerade durch Minimierung der Summe der Fehlerquadrate berechnet.

```python
# Datenpunkte erzeugen
np.random.seed(17)  # Für die Reproduzierbarkeit der Zufallsdaten
x = np.linspace(0, 10, 100)  # Beispielwerte für die Prädiktorvariable
y = 2 * x + np.random.normal(0, 2, 100)  # Beispielwerte für die Zielvariable (mit zufälligem Fehler)

# Definiere das lineare Modell
def linear_func(params, x):
    m, b = params
    return m * x + b

# Definiere die Fehlerfunktion (quadratischer Fehler)
def error_func(params,x):
    y_pred = linear_func(params, x)
    return np.sum((y_pred - y)**2)

# Wähle einen Startpunkt für den Algorithmus
initial_guess = [0.0, 0.0]

# Minimiere den quadratischen Fehler
result = minimize(error_func, initial_guess, args=(x,))


# Extrahiere die geschätzten Parameter der Regressionsgeraden
m, c = result.x

# Gib die Regressionsgerade aus
print("Die Regressionsgerade hat die Gleichung: y = {:.2f}x + {:.2f}".format(m, c))
```

Anschließend plotten wir die Datenpunkte und die gefundene Gerade.

```python
# Plot der Daten und der Regressionsgeraden
plt.figure(figsize=(12, 6))
plt.scatter(x, y, label='Datenpunkte')
plt.plot(x, linear_func([m, c], x), color='red', label='Regressionsgerade')
plt.xlabel('x', fontsize = 22)  
plt.ylabel('f(x)',  fontsize = 22)

# Skalenwerte auf Achsen vergrößern
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.legend(fontsize=22)

plt.grid(True)
plt.show()
```
