# Lineare Regression


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize


# Datenpunkte erstellen
np.random.seed(17)  # Für die Reproduzierbarkeit der Zufallsdaten
x = np.linspace(0, 10, 100)  # Beispielwerte für die Prädiktorvariable
y = 2 * x + np.random.normal(0, 2, 100)  # Beispielwerte für die Zielvariable (mit Rauschen)

# Definiere die lineare Funktion
def linear_func(params, x):
    m, c = params
    return m * x + c

# Definiere die Fehlerfunktion (quadratischer Fehler)
def error_func(params,x):
    y_pred = linear_func(params, x)
    return np.sum((y_pred - y)**2)

# Wähle einen Startpunkt für den Algorithmus
initial_guess = [0.0, 0.0]

# Minimiere den quadratischen Fehler
result = minimize(error_func, initial_guess, args=(x,))


# Extrahiere die geschätzten Parameter der Regressionsgeraden
m, c = result.x

# Gib die Regressionsgerade aus
print("Die Regressionsgerade hat die Gleichung: y = {:.2f}x + {:.2f}".format(m, c))

# Plot der Daten und der Regressionsgeraden
plt.scatter(x, y, label='Datenpunkte')
plt.plot(x, linear_func([m, c], x), color='red', label='Regressionsgerade')
# Achsentitel und Legende hinzufügen
plt.xlabel('Prädiktorvariable')
plt.ylabel('Zielvariable')
plt.title('Lineare Regression')
plt.legend()
plt.grid(True)
plt.show()
```
